```julia
dir=Dates.format(now(), "yyyy-mm-dd HH:MM:SS")
using JLD

include("equilibration.jl")
include("production.jl")
include("Ising.jl")
include("ising_over_temp.jl")

n_grid = 61
J = 1.0
L = 200000000
Tmin = 2.269
Tinc = 1.0
Tmax = 2.269
P = 4
r = 1:1:convert(Int,floor(n_grid/2))
Ts, len = ising_over_temp(n_grid,J,L,Tmin,Tinc,Tmax)
corr_avg, SiSj_avg, err, Ts = Ising(n_grid,L,J,P,Tmin,Tinc,Tmax,len,Ts)

#File = (r[:,:]corr_avg[1,1,:])

mkpath("Data/$n_grid/$dir")
fid=open("Data/$n_grid/$dir/r_corrfunc_err.dat","w")

for i in 1:size(r[:,:])[1]-1
    if SiSj_avg[1,1,i] > 0
    @printf(fid,"%6f %16f %20f\r\n",log(r[:,:][i]),log(SiSj_avg[1,1,i]),err[1,1,i]);
    end
end
#@printf(fid,"%s\r\n\r\n\r\n\r\n","");
close(fid);
fid=open("Data/$n_grid/$dir/sys_details.dat","w")
@printf(fid,"%s %s\r\n","Date/Time : ",Dates.format(now(), "yyyy-mm-dd HH:MM:SS"));
@printf(fid,"%s %f\r\n","J=",J);
@printf(fid,"%s %u\r\n","Lattice Size=",n_grid);
@printf(fid,"%s %f\r\n","Minimum temperature = ",Tmin);
@printf(fid,"%s %f\r\n","Maximum temperature = ",Tmax);
@printf(fid,"%s %f\r\n","Increment in temperature = ",Tinc);
@printf(fid,"%s %u\r\n","No.of Production run = ",P);
@printf(fid,"%s %u\r\n","No.of steps in production run = ",L);
@printf(fid,"%s\r\n","Data:");
@printf(fid,"%6s %12s %14s\r\n\r\n","r (distance)","Correlation function","Error");
close(fid)
using Gadfly
plot(x=r[:,:],y=corr_avg[1,1,:],Guide.xlabel("r"),Guide.ylabel("(SiSj)-(Si)(Sj)"), Geom.point, Geom.line)
```
